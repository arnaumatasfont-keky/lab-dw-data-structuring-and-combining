In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
import pandas as pd
url_chall2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_chall2 = pd.read_csv(url_chall2)
```

```python
if 'unnamed:_0' in df_chall2.columns:
    df_chall2 = df_chall2.drop(columns=['unnamed:_0'])

df_chall2['effective_to_date'] = pd.to_datetime(df_chall2['effective_to_date'])

df_chall2['number_of_open_complaints'] = df_chall2['number_of_open_complaints'].astype(int)

df_chall2 = df_chall2.drop_duplicates()

df_chall2.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_revenue = df_chall2.pivot_table(index='sales_channel', 
                                      values='total_claim_amount', 
                                      aggfunc='sum')


pivot_revenue = pivot_revenue.round(2)


pivot_revenue = pivot_revenue.sort_values(by='total_claim_amount', ascending=False)


display(pivot_revenue)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_clv = df_chall2.pivot_table(index=['gender', 'education'], 
                                  values='customer_lifetime_value', 
                                  aggfunc='mean')


pivot_clv = pivot_clv.round(2)


display(pivot_clv)
```
